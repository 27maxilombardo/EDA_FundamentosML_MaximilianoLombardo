```python
# Importa las bibliotecas necesarias
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Descarga y carga el dataset
path = kagglehub.dataset_download("pavansubhasht/ibm-hr-analytics-attrition-dataset")
ruta_dataset = path + '/WA_Fn-UseC_-HR-Employee-Attrition.csv'
df = pd.read_csv(ruta_dataset)

# Exploración inicial del dataset
print("La ruta donde se guardó el dataset es:", path)
print("\nUn vistazo a las primeras filas del dataset:")
display(df.head())  # Usando 'display' para una mejor visualización en Colab
print("\nInformación sobre las columnas del dataset:")
print(df.info())
print("\nEstadísticas descriptivas del dataset:")
print(df.describe())

# Análisis de variables específicas
print("\nSatisfacción laboral:")
print(df['JobSatisfaction'].value_counts())
print(df['JobSatisfaction'].mean())
print("\nNivel de ingreso:")
print(df['MonthlyIncome'].describe())
print(df.sort_values(by=['MonthlyIncome'], ascending=False).head())

# Visualización de datos
plt.figure(figsize=(12, 8))  # Ajusta el tamaño de la figura

# Gráfico de barras para 'Attrition'
plt.subplot(2, 2, 1)  # Crea un subplot dentro de la figura
sns.countplot(x='Attrition', data=df)
plt.title('Distribución de Attrition')

# Gráfico de pastel para 'Department'
plt.subplot(2, 2, 2)
df['Department'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribución de Department')

# Histograma para 'Age'
plt.subplot(2, 2, 3)
plt.hist(df['Age'], bins=10)
plt.title('Distribución de Age')
plt.xlabel('Age')
plt.ylabel('Frecuencia')

# Diagrama de caja para 'MonthlyIncome'
plt.subplot(2, 2, 4)
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('MonthlyIncome por Attrition')

plt.tight_layout()  # Ajusta el espaciado entre los subplots
plt.show()

# Gráfico de dispersión para 'TotalWorkingYears' vs 'MonthlyIncome'
plt.figure(figsize=(8, 6))  # Crea una nueva figura para este gráfico
plt.scatter(df['TotalWorkingYears'], df['MonthlyIncome'])
plt.title('TotalWorkingYears vs MonthlyIncome')
plt.xlabel('TotalWorkingYears')
plt.ylabel('MonthlyIncome')
plt.show()
```
